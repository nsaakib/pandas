```python
deth_over =delivery[delivery['over']>15]
```

```python
all_batsman = deth_over.groupby('batter')['batsman_runs'].count()
all_batsman
```

This code calculates how many times each batsman has batted. Here's a breakdown:

deth_over.groupby('batter'): Groups the data by the 'batter' column, meaning it organizes all entries for each individual batsman.
['batsman_runs'].count(): Counts the number of entries (or balls faced) for each batsman within the group.
The result, stored in all_batsman, is a series that shows each batsman's name and their total number of balls faced.
In short, it counts the total number of balls faced by each batsman.

```python
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

This code filters the list of batsmen who have faced more than 200 balls. Here's what it does:

x = all_batsman > 200: Creates a boolean series where each entry is True if the batsman's total balls faced is more than 200, otherwise False.

batsman_list = all_batsman[x].index.tolist(): Selects only the batsmen with more than 200 balls faced and converts their names (index) into a list.

In short, it creates a list of batsmen who have faced over 200 balls.

##### strike rate calculation
Runs scored by all these batsman

Balls played by all these batsman

```python
final = deth_over[deth_over['batter'].isin(batsman_list)]
```

This code filters the data to include only the records where the batter's name is in the list batsman_list.

deth_over['batter'].isin(batsman_list): Checks if each 'batter' in the data is in the list batsman_list.

deth_over[...]: Selects only the rows where the condition is true.
    
In short: It creates a new dataset final containing only the records of batsmen who faced more than 200 balls.

```python
runs = final.groupby('batter')['batsman_runs'].sum()
```

This code calculates the total runs scored by each batsman in the filtered data.

final.groupby('batter'): Groups the data by each batsman.

['batsman_runs'].sum(): Sums up all the runs scored by each batsman.

In short: It provides the total runs scored by each batsman in the final dataset

```python
balls = final.groupby('batter')['batsman_runs'].count()
```

This code calculates the total number of balls faced by each batsman in the filtered data.

final.groupby('batter'): Groups the data by each batsman.

['batsman_runs'].count(): Counts how many times each batsman appears, which corresponds to the number of balls faced.
